```python
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
life_expectancy = pd.read_csv('../data/Life Expectancy Data.csv')
life_expectancy
```

```python
# deleting extra spaces
life_expectancy.rename(columns = {'Life expectancy ': 'Life expectancy',
                                  'Measles ': 'Measles',
                                  ' BMI ': 'BMI',
                                  'under-five deaths ': 'under-five deaths',
                                  'Diphtheria ': 'Diphtheria',
                                  ' thinness  1-19 years':  'thinness  1-19 years',
                                  ' thinness 5-9 years': 'thinness 5-9 years'}, inplace = True)
life_expectancy.columns
```

```python
# check for missing values
missing_values = life_expectancy.isna().sum()
missing_values
```

```python
# handelling missing values
life_expectancy = life_expectancy.dropna(subset = ['Life expectancy', 'Adult Mortality'])

cols_with_na = life_expectancy.isna().sum()[life_expectancy.isna().sum() > 0].index.tolist()

for column in cols_with_na:
    mean_value = life_expectancy.groupby('Country')[column].transform('mean')
    life_expectancy.loc[:, column] = life_expectancy[column].fillna(mean_value)
```

```python
missing_values = life_expectancy.isna().sum()
missing_values
```

```python
# Handling missing values for columns that still have NA entries
cols_with_na = life_expectancy.isna().sum()[life_expectancy.isna().sum() > 0].index.tolist()

for column in cols_with_na:
    mode_value = life_expectancy[column].mode()[0] 
    life_expectancy.loc[:, column] = life_expectancy[column].fillna(mode_value)
```

```python
missing_values = life_expectancy.isna().sum()
missing_values
```

```python
# mapping for ordinal variables
label_encoder = LabelEncoder()
life_expectancy.loc[:, 'Status'] = label_encoder.fit_transform(life_expectancy['Status']).astype('int32')

life_expectancy
```

```python
# dropping the country column
life_expectancy = life_expectancy.drop(columns = ['Country'])
```

```python
# Standard Scaling
X_train, X_test, y_train, y_test = train_test_split(life_expectancy.drop(columns = ['Life expectancy']), 
                                                    life_expectancy['Life expectancy'], 
                                                    test_size = 0.2)

features = X_train.columns.tolist()

std_scaler = StandardScaler()
X_train[features] = X_train[features].astype('float64')
X_test[features] = X_test[features].astype('float64')

X_train[features] = std_scaler.fit_transform(X_train[features])
X_test[features]  = std_scaler.transform(X_test[features])
```

```python
# generating polynomial features
poly_transformer = PolynomialFeatures(degree = 2)
poly_transformer.fit(X_train)

X_train_poly = poly_transformer.transform(X_train)
X_test_poly = poly_transformer.transform(X_test)
```

```python
# training the model
model = LinearRegression()
model.fit(X_train_poly, y_train)
```

```python
# predicting on the train and test sets
y_pred_train = model.predict(X_train_poly)
y_pred_test = model.predict(X_test_poly)
```

```python
# evaluate your model
from sklearn.metrics import r2_score

r2_train = r2_score(y_train, y_pred_train)
r2_test = r2_score(y_test, y_pred_test)

print("R^2 Score on Training Set:", r2_train)
print("R^2 Score on Test Set:", r2_test)
```
